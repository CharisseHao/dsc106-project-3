```python
import pandas as pd

# Step 1: Read JSON data
with open('yelp_academic_dataset_business.json', 'r') as file:
    json_data = file.read()

# Step 2: Convert to DataFrame
data_df = pd.read_json('yelp_academic_dataset_business.json', lines=True)
```

```python
cleaned = data_df[data_df['is_open'] == 1][['business_id', 'state', 'stars', 'categories']]
cleaned
```

```python
def categorize(x):
    if x is None:
        return 'Other'
    x = x.lower()
    if ('restaurants' in x) or ('food' in x) or ('bar' in x):
        return 'Restaurants'
    elif ('shopping' in x) or ('fashion' in x):
        return 'Shopping'
    elif 'home services' in x:
        return 'Home Services'
    elif ('beauty & spas' in x) or ('hair salons' in x):
        return 'Beauty & Spas'
    elif 'health & medical' in x:
        return 'Health & Medical'
    elif ('automotive' in x) or ('auto repair' in x):
        return 'Automotive'
#     elif ('bars') (and nightlife
    elif 'hotel' in x:
        return 'Hotels & Travel'
    else:
        return 'Other'
```

```python
cleaned['category'] = cleaned['categories'].apply(categorize)
cleaned.head(5)
```

```python

# Step 3: Save as CSV

# data_df.to_csv('yelp_academic_dataset_business.csv', index=False)
cleaned.to_csv('cleaned_yelp_dataset_business.csv', index=False)

print("Conversion completed. CSV file saved successfully.")
```
